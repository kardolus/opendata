```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the file
df = pd.read_csv("../data/ferry_sb.csv")

# Group by the 'Stop' column and get the average boardings
average_boardings = df.groupby('Stop')['Boardings'].mean().sort_values(ascending=False)

# Plot the data using a bar plot
plt.figure(figsize=(12,6))
average_boardings.plot(kind='bar', color='skyblue')
plt.title("Average Number of Boardings by Stop")
plt.ylabel("Average Boardings")
plt.xlabel("Stop")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the file
df = pd.read_csv("../data/ferry_sb.csv")

# Group by the 'Stop' column and count the number of occurrences for each stop
stop_counts = df['Stop'].value_counts().sort_values(ascending=False)

# Plot the data using a bar plot
plt.figure(figsize=(12,6))
stop_counts.plot(kind='bar', color='lightgreen')
plt.title("Frequency of Ferry Stops by Station")
plt.ylabel("Number of Stops")
plt.xlabel("Station")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
